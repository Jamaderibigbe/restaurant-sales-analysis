**Q.2) Most Selling Product?**

- By Quantity
- By Revenue

```python
data.head()
```

```python
# Using the group by function

data.groupby('Product')['Quantity'].sum()
```

```python
# to sort these values

data.groupby('Product')['Quantity'].sum().sort_values(ascending = False)
```

```python
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending = False)
most_quantity
```

```python
type(most_quantity)
```

```python
most_quantity = most_quantity.reset_index()        # to convert the index of a series into a column to form a dataframe
most_quantity
```

```python
type(most_quantity)
```

```python
# ploting a bar graph 

plt.figure(figsize = (9, 4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color = ['red', 'black', 'green', 'yellow', 'cyan'], width = 0.5) ;
plt.title("Most Selling Product - By Quantity")
plt.xlabel("Product")
plt.ylabel("Quantity") ;
```

**By Revenue**

```python
data.head()
```

```python
# adding new column to the dataset

data["Revenue"] = data["Price"] * data["Quantity"]
```

```python
data.head()
```

```python
data.groupby('Product')['Revenue'].sum().sort_values(ascending = False)
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending = False)

most_revenue
```

```python
most_revenue = most_revenue.reset_index()

most_revenue
```

```python
plt.figure(figsize = (9, 4) )
plt.bar(most_revenue['Product'], most_revenue['Revenue'], color = ['green', 'red', 'black', 'yellow', 'cyan'], width = 0.3) ;
plt.title("Most Selling Product - By Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue") ;
```
